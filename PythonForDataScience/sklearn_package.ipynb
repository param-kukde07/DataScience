```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
```

```python
N = 30
x = np.arange(N)
y = 4*x-7 + (10*np.random.randn(N))
plt.scatter(x, y)
```

```python
x.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)
model.fit(x[:, np.newaxis],y)
yfit = model.predict(x[:, np.newaxis])
plt.scatter(x, y)
plt.plot(x, yfit,color='red')
```

```python
from sklearn.datasets import make_blobs
X , y = make_blobs(n_samples=50000, centers=3, cluster_std=1.2)
plt.scatter(X[:,0], X[:,1], c=y, cmap="autumn");
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33)
```

```python
from sklearn.svm import SVC
```

```python
md1 = SVC(gamma = "auto")
md1.fit(X_train, y_train)
```

```python
y_predict = md1.predict(X_test)
```

```python
acc = 100*np.sum(y_predict == y_test)/y_test.size
print("Accuracy is: ", acc,"%")
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
md1 = RandomForestClassifier(n_estimators = 20)
md1.fit(X_train, y_train)
```

```python
y_predict = md1.predict(X_test)
acc = 100*np.sum(y_predict == y_test)/y_test.size
print("Accuracy is: ", acc,"%")
```
